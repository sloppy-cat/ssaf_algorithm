#  연습문제 1 gravity
100 * 100 방에 상자들이 쌓여있다.
이때 방이 오른쪽으로 90도 기울어 상자들이 쏟아지게 되었을 때 가장 많이 떨어지는 상자의 낙차는 얼마인가?


```python
def build_data(): 
    data = []
    for i in range (0,10):
        data += [int(input())]
    return data
data = build_data()

nak = 0
for i in range(0,9):
    new_nak = 9-i
    if nak >= new_nak:
        break
    for j in range (i+1,9):
        if data[i] <= data[j]:
            new_nak -= 1
    if new_nak > nak:
        nak = new_nak
print(nak)
```

# Baby-gin
0~9 카드중 6장 뽑을때 3장이상 연속 = run, 3장이상 같으면 triplet
6장의 카드가 run과 triplet이라면 baby-gin

baby-gin 여부 판단 프로그램.

```python
import copy
card = list(map(int,input().split()))
card.sort()
set_card = copy.copy(set(card))
if len(set_card) == 1:
    print('True')
for x in set_card:
    if card.count(x) >= 3:
        for i in range(3):
            card.remove(x)

    if len(card) != 0:
        a = len(card)
        for i in range (a//3,0,-1):
            if card[i*3-1] == card[i*3-2]+1 and card[i*3-3]+1 == card[i*3-2]:
                for i in range(3):
                    card.pop()
if len(card) == 0:
    print('True')
else:
    print('False')
```

# workshop jomang

```python
jomang = 0
N = int(input())
datas = list(map(int,input().split()))

for i in range(2,N-2):
    jo = datas[i] - max(datas[i-2],datas[i-1],datas[i+1],datas[i+2])
    if jo>0:
        jomang += jo
#result += [jomang]
print(jomang)
```

```python
def electronic_bus():
    T = int(input())
    for test_case in range(1, T + 1):
        K, N, M = list(map(int,input().split()))
        bus = [0]*(N+K)
        charger = list(map(int,input().split()))
        
        for x in charger:
            bus[x] = 1
        result = 0
        i = 0
        while i<(N):
            for j in range ((i+K),i,-1):
                if j >= N :
                    i = j
                    break
                if bus[j] == 1:
                    i = j
                    result += 1
                    break
        print(f'#{test_case} {result}')
electronic_bus()
```

# flatten
덤프(젤 높은거 젤 낮은데로 옮기는 작업) 횟수가 제한되어있을때 들어온 케이스의 최고점 , 최저점 차이 구하라

```python
def flatten():
    T = int(input())
    for test_case in range(1, T + 1):
        N, M = map(int,input().split())
        ai = list(map(int,input().split()))      

        minmin = 1000000
        maxmax = 0
        for i in range(0,N-M+1):
            if sum(ai[i:i+3])> maxmax:
                maxmax = sum(ai[i:i+3])
            elif sum(ai[i:i+3]) < minmin:
                minmin = sum(ai[i:i+3])
        print(f'#{test_case} {maxmax-minmin}')
```

```python
import copy
T = int(input())
for test_case in range(1, T + 1):
    N, K = list(map(int,input().split()))

    arrr = []
    arr = [0]*12
    while True:
        for i in range(len(arr)-1,-1,-1):
            if arr[i] == 1:
                arr[i] = 0
            else :
                arr[i] = 1
                break

        if arr.count(1) == N:
            arrr.append(copy.copy(arr))
        elif arr.count(1) == 12:
            break
    r = 0
    for ar in arrr:
        check = 0
        for i in range(len(ar)):
            if ar[i] == 1:
                check += (i+1)
        if check == K:
            r += 1
    print(f'#{test_case} {r}')
```

```python
arr = [1,2,3,4,5,6]
n = len(arr)
for i in range(1<<n):
    for j in range(n+1):
        if i &(1<<j):
            print(arr[j], end=", ")
        print(f'    {bin(i)} --- {bin(1<<j)}    ')
    print()
print()
```

```python
def find_bibi(l,r,F,n):
    c = int((l+r)/2)
    if c == F:
        return n
    elif c > F:
        return find_bibi(l,c,F,n+1)
    else :
        return find_bibi(c,r,F,n+1)
def binary():
    T = int(input())
    for test_case in range(1, T + 1):
        P, A, B = list(map(int,input().split()))
        l,r = 1,P
        a = find_bibi(l,r,A,0)
        b = find_bibi(l,r,B,0)
        R = 0
        if a>b:
            R = 'B'
        elif b>a:
            R = 'A'
        print(f'#{test_case} {R}')
#color()
#sub_set()
binary()
```
